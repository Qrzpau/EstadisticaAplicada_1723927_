```python
#repaso rápido pandas
#siempre el primer paso es llamar las librerias que utilizaremos
import pandas as pd
import numpy as np
```

```python
## Importar los datos a utilizar
calificaciones = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/Calificaciones.csv')
```

```python
calificaciones.info()
```

```python
## Renombrar
calificaciones= calificaciones.rename(columns={'Calficación Final': 'CalFinal'})
```

```python
calificaciones
```

```python
# Con loc tomamos las demas columnas excepto una a especificar
calificaciones= calificaciones.loc[:,calificaciones.columns!='Id']
```

```python
calificaciones
```

```python
#primeras 5 filas
calificaciones.head()
```

```python
#Promedio del grupo
calificaciones['CalFinal'].mean()
```

```python
## alumnos
calificaciones['Nombre'].count()
```

```python
## Aprobados y reprobados
calificaciones['Situación']=np.where(calificaciones['CalFinal']>=70, 'Aprobado', 'Reprobado')
```

```python
calificaciones.head()
```

```python
## Promedio de aprobados y el promedio de reprobados
calificaciones.groupby('Situación').mean().round(1)
```

```python
## aggregate agg
# contar aprobados y reprobados
## se pueden pedir mas caracteristicas con aggregate
calificaciones.groupby('Situación')['CalFinal'].agg(['count', 'mean', 'std']).round(1)
```

## Matplotlib

Es el graficador estándar de Python

```python
## importar libreria

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Creamos una lista que se llamará edades

personas = [1, 2, 3, 4]
edades = [28, 21, 30, 27]
```

```python
###
plt.scatter(personas,edades)
## quitar codigo de encima de forma correcta con .show
plt.show()

##Tambien se puede quitar con punto y coma
#plt.plot(personas,edades);

# graficas de barras
plt.bar(personas, edades);
```

```python
## para personalizar el grafico
plt.bar(personas,edades, color=['blue', 'red', 'yellow','green'])
plt.xlabel('personas')
plt.ylabel('edad (años)')
plt.title('Edades de 4 desconocidos')
plt.show()
```
