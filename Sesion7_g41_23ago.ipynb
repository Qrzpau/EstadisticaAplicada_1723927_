```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import statsmodels.formula.api as smf
```

```python
#datos
datos = pd.read_excel('modelolineal.xlsx')
```

```python
datos = datos[['Y', 'X']]
```

```python
datos
```

```python
#Hacemos gráfico de dispersión
plt.scatter(datos['X'], datos['Y'])
plt.title('Gráfico de dispersión de X y Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show ()
```

```python
#Aparentemente existe una relación lineal entonces se puede estimar minimos cuadrados ordinarios
#Se puede estimar MCO .
#OJO: SIMPRE HACE UN ANALISIS VISUAL PARA DETERMINAR SI ES NECESARIO ESTIMAR LOS MINIMOS CUADRADOS
```

```python
reg = smf.ols(formula = 'Y~X', data = datos)
#Qué y cómo tiene que estimar la regresión
results = reg.fit()
#Aqui le decimos que lo estime
print (results.summary())
#Mostrar el resultado de la regresion
```
