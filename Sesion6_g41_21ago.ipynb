```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nacimientos = pd.DataFrame({'municipio':['Mty', 'Apodaca', 'Escobedo', 'Santiago'], 'bb_nacidos' : [233, 569, 780, 123], 'moridos' : [123, 80, 23, 12]})
```

```python
nacimientos = nacimientos.sort_values('bb_nacidos', ascending=False)
nacimientos
```

```python
plt.bar(nacimientos['municipio'],nacimientos['bb_nacidos'])
plt.title('Nacimientos en N.L en 2020')
plt.xlabel('municipios')
plt.ylabel('Cantidad de bebes')
plt.show()
```

```python
## Creamos una lista que se llamará edades

personas = ['Juan','Pedro','Ana','Ale']
edades = [23,18,20,31]
```

```python
###
plt.scatter(personas,edades)
## quitar codigo de encima de forma correcta con .show
plt.show()

##Tambien se puede quitar con punto y coma
#plt.plot(personas,edades);

# graficas de barras
plt.bar(personas, edades);
```

```python
## para personalizar el grafico
plt.bar(personas,edades, color=['blue', 'red', 'yellow','green'])
plt.xlabel('personas')
plt.ylabel('edad (años)')
plt.title('Edades de 4 desconocidos')
plt.show()
```

```python
#fig es el nombre del lienzo
fig, (panel1, panel2) = plt.subplots(2,1)
fig.tight_layout()

##tight es para que se estiran similares

#gráfica1 nacimientos
x1 = nacimientos['municipio']
y1 = nacimientos['bb_nacidos']
panel1.bar(x1,y1)

#gráfica2 edades

x2 = personas
y2 = edades
panel2.bar(x2,y2)
plt.show()
```
